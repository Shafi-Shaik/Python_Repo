```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Chart

```python
Ages=np.sort( np.random.randint(10,70,12) ) #12 values between 10 and 70 in ascending order
weight=np.sort( np.random.randint(45,120,12))[::-1]#12 values between 45 and 120 in descending order

Ages, weight
```

```python
plt.plot(Ages,weight,ls='dashed',lw=3,c='m')



plt.xlim(0,85)
plt.ylim(40,130)


plt.title(' Ages vs Weight Distribution of 12 people')
plt.xlabel('Ages')
plt.ylabel('Weight')

plt.show()
```

## Histogram

```python
Ages=[11,31,33,45,34,60,71,75,87,89,90,52,45,10,87]
weight=[14,45,55,67,78,76,71,40,87,89,90,52,45,45,87]

age_bins= [10,30,70,100]
wt_bins= [10,50,80,100,150]

plt.figure(figsize=(12, 4))
            #subplot 1 (row=1, column=2, index=1)
    
plt.subplot(1,2,1)
plt.hist(Ages,age_bins,histtype='stepfilled',color='g')

plt.title('No. of people w.r.t Age bins')
plt.xlim(0,120)
plt.xticks(age_bins)



              #subplot 2 (row=1, column=2, index=2)
plt.subplot(1,2,2)
plt.hist(weight,wt_bins,histtype='stepfilled',orientation='vertical',color='m')

plt.title('No. of people w.r.t weight bins')
plt.xlim(0,150)
plt.xticks(wt_bins)

#Super Title
plt.suptitle(' This is Super title', fontdict={'color':'#00aeff','size':40,'family':'calibri'})


plt.show()
```

## Bar Chart

```python
# Data Prep
marks = [  92,  42,  61, 57,  18,  46]
subjects = ['Maths','Science','Social','Hindi','Urdu','English']

clrs=['g','y','m','r']
```

```python
plt.figure(figsize=[5,8]) # horizontal=5,vertical=8


#Subplot 1
plt.subplot(2,1,1)
plt.bar(subjects,marks,color=clrs)
plt.title('Vertical View')

#Subplot 2
plt.subplot(2,1,2)
plt.barh(subjects,marks,color=clrs)
plt.title('Horizontal View')

plt.suptitle('Marks secured by Ravi in Quarterly Exams',size=15)

plt.show()
```

## Grouped Bar Chart

```python
# Data Prep

subjects = ['Maths','Science','Social','Hindi','Urdu','English']
ravi_marks = [  92,  42,  61, 57,  18,  46]
irfan_marks = [  34,  45,  67, 57,  67,  34]
andrew_marks = [  72,  42,  46, 66,  55,  34]

#bar chart properties
x = np.arange(len(subjects))  # the label locations
wd = 0.2
```

```python
plt.figure(figsize=[12,6])

# plt.subplots() is a function that returns a tuple containing a figure and axes object(s). 
# Thus when using fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax. 
# Having fig is useful if you want to change figure-level attributes or save the figure 
# as an image file later (e.g. with fig.savefig('yourfilename.png')
fig,ax=plt.subplots()

# we need to increase x value as these charts are indirectly pointing (x,y) values.
# for ravi if (1,y) with bar-width 0.2 then irfan should have (1.2,y) and andrew (1.4,y)

bar1=ax.bar(x,ravi_marks,wd,label='Ravi')
bar2=ax.bar(x+wd,irfan_marks,wd,label='Irfan')
bar3=ax.bar(x+2*wd,andrew_marks,wd,label='Andrew')


ax.set_title('Marks Scored by 3 students in Quarterly Exam')
ax.set_xlabel('Subjects')
ax.set_ylabel('Marks')

# resetting x ticks to subjects name
ax.set_xticks(x+wd,subjects) # x+wd to align name to the exact center of the 3 charts
ax.legend()
ax.grid(axis='y')

plt.show()
```

## Stacked Bar Chart

```python
# Data Prep

subjects = ['Maths','Science','Social','Hindi','Urdu','English']
ravi_marks = np.array([  92,  25,  61, 75,  18,  46])
irfan_marks = np.array([  34,  45,  67, 57,  67,  34])
andrew_marks = np.array([  72,  42,  46, 66,  55,  34])

wd = 0.2
```

```python
fig,ax=plt.subplots()

bar1=ax.bar(subjects,ravi_marks,wd,label='Ravi')
bar2=ax.bar(subjects,irfan_marks,wd,label='Irfan',bottom=ravi_marks)
bar3=ax.bar(subjects,andrew_marks,wd,label='Andrew',bottom=ravi_marks+irfan_marks)


ax.set_title('Marks Scored by 3 students in Quarterly Exam')
ax.set_xlabel('Subjects')
ax.set_ylabel('Marks')

ax.legend()
ax.grid(axis='y')

plt.show()
```

##### How Age affects the Speed of the Car

```python

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
plt.scatter(x, y, color = 'hotpink',label='day1')


#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.scatter(x, y, color = '#88c999',label='day2')
plt.colorbar() # for right side color gradient bar (0.0- 1.0)

plt.legend()
plt.show()
```

## Pie Chart

```python
population = [4650,1384,745,504,438,51]
continents = ['Asia','Africa','Europe','Australia','North America','South America']
```

```python
plt.figure(figsize=[8,8])

plt.pie(population,labels=continents, explode=[0.1,0,0,0,0,0],autopct='%1.2f%%',startangle=90)


plt.title("Population by Continent - 2020")
plt.show()
```
