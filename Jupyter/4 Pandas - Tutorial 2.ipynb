<h2 style="text-align:center; font-weight:bold;color:Green">      
     Pivot and Unpivot
</h2>

```python
pivot_df=pclass_sex_mean_age.pivot_table(values='age',
                                         index='sex',
                                         columns='pclass', 
                                         aggfunc =np.mean
                                        )

# pclass                1             2                   3
# sex
# female            34.611765         28.722973            21.750000
# male              41.281386         30.740707            26.507589

# pivot_df=pivot_df.reset_index() # reset_index() will remove multi-index to single index
# pclass    sex        1          2           3  
# 0        female   34.611765   28.722973   21.750000
# 1        male    41.281386    30.740707   26.507589


pivot_df=pivot_df.reset_index().rename_axis(columns=None) # rename_axis() removes pclass as row_index name
#         sex        1          2           3  
# 0      female   34.611765   28.722973   21.750000
# 1       male    41.281386    30.740707   26.507589

pivot_df
```

```python
unpivot_df = pivot_df.melt(var_name=['pclass'],
                           value_name='avg_age',                           
                           col_level=0,
                           id_vars =['sex']
                         )
unpivot_df
```

<h2 style="text-align:center; font-weight:bold;color:Green">      
     Missing Values
</h2>

<h6 style="text-align:left; font-weight:bold;color:blue">      
    1. Finding Missing Values in each column
</h6>

```python
# isnull() and isna()
df=titanic.copy()
df.isna().sum()
```

<h6 style="text-align:left; font-weight:bold;color:blue">      
    2. Finding rows with atleast one column with Null value
</h6>

```python
missings = df[df.isnull().any(axis=1)]
missings.head(3)
```

<h6 style="text-align:left; font-weight:bold;color:blue">      
    3. Finding Mean and Median difference in order to impute:
</h6>

we will use **_Mean_** if there is no significant differnce b/w Mean & Median
else we will use **_Median_**

```python
df.describe()
```

<h6 style="text-align:left; font-weight:bold;color:blue">      
    4. Imputing with Mean value for 'Age' column as there is no difference in mean & median of age
</h6>

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

<h6 style="text-align:left; font-weight:bold;color:blue">      
    5.  Now 'Age' does not have Null values. Now we will impute cabin and embarked columns
</h6>


```python
df.isna().sum()
```


Since cabin & embarked columns are categorical columns, we will use mode(most repititive value) of respective columns

```python
df['cabin'].value_counts() # getting number of times each value has repeated
```

<h6 style="text-align:left; font-weight:bold;color:blue">      
    6. Imputing with Mode value for 'Cabin' & 'Embarked' column
</h6>

```python
# fillna() - fill all null values
# .index[0] - get the first value 
df['cabin'].fillna(df['cabin'].value_counts().index[0],inplace=True)
```

```python
df['embarked'].fillna(df['embarked'].value_counts().index[0],inplace=True)
```

```python
df.isna().sum()
```

<h2 style="text-align:center; font-weight:bold;color:Green">      
     Short cut for Imputing in one single code using Lambda 
</h2>

```python
df2=titanic
df2.isna().sum()
```

```python
df2=df2.apply(lambda x:x.fillna(x.mean() if x.dtype=='float' \
                                else x.fillna(x.value_counts().index[0])))
```

```python
df2.isna().sum()
```
